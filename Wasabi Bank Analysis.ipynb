```python
# Question 1
def map_location_to_bin(location_id):
    if location_id.startswith('EN'):
        return 'Enugu'
    elif location_id.startswith('FC'):
        return 'FCT'
    elif location_id.startswith('LA'):
        return 'Lagos'
    elif location_id.startswith('KN'):
        return 'Kaduna'
    elif location_id.startswith('RI'):
        return 'Rivers'
    else:
        return 'Other'

transactions_df['LocationBin'] = transactions_df['LocationID'].apply(map_location_to_bin)

# Group transactions by location bin and transaction type and calculate the average transaction amount
avg_amount_by_location_type = transactions_df.groupby(['LocationBin', 'TransactionTypeID'])['TransactionAmount'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_amount_by_location_type, x='TransactionTypeID', y='TransactionAmount', hue='LocationBin')
plt.title('Average Transaction Amount by Location and Transaction Type')
plt.xlabel('Transaction Type')
plt.ylabel('Average Transaction Amount')
plt.show()
```

```python
def map_location_to_bin(location_id):
    if location_id.startswith('EN'):
        return 'Enugu'
    elif location_id.startswith('FC'):
        return 'FCT'
    elif location_id.startswith('LA'):
        return 'Lagos'
    elif location_id.startswith('KN'):
        return 'Kaduna'
    elif location_id.startswith('RI'):
        return 'Rivers'
    else:
        return 'Other'

transactions_df['LocationBin'] = transactions_df['LocationID'].apply(map_location_to_bin)


avg_amount_by_location_type = transactions_df.groupby(['LocationBin', 'TransactionTypeID'])['TransactionAmount'].mean().reset_index()


plt.figure(figsize=(10, 6))
sns.barplot(data=avg_amount_by_location_type, x='LocationBin', y='TransactionAmount', hue='TransactionTypeID')
plt.title('Average Transaction Amount by Location and Transaction Type')
plt.xlabel('Location')
plt.ylabel('Average Transaction Amount')
plt.show()
```

```python
#question 2
transactions_df['TransactionStartDateTime'] = pd.to_datetime(transactions_df['TransactionStartDateTime'])
transactions_df['TransactionDay'] = transactions_df['TransactionStartDateTime'].dt.date
transactions_df['TransactionHour'] = transactions_df['TransactionStartDateTime'].dt.hour

daily_transactions = transactions_df.groupby(['LocationID', 'TransactionDay', 'TransactionHour']).size().reset_index(name='TransactionCount')
max_daily_transactions = daily_transactions.groupby('LocationID')['TransactionCount'].max()
max_location = max_daily_transactions.idxmax()
max_location_name = location_df[location_df['LocationID'] == max_location]['Location Name'].values[0]

print(f"ATM Location with the highest daily transactions: {max_location_name}")

```

```python
transactions_df['TransactionStartDateTime'] = pd.to_datetime(transactions_df['TransactionStartDateTime'])

transactions_df['TransactionHour'] = transactions_df['TransactionStartDateTime'].dt.hour

hourly_transaction_counts = transactions_df['TransactionHour'].value_counts().sort_index()

most_transactions_hour = hourly_transaction_counts.idxmax()
most_transactions_count = hourly_transaction_counts.max()


plt.figure(figsize=(10, 6))
plt.bar(hourly_transaction_counts.index, hourly_transaction_counts.values)
plt.xlabel('Hour of the Day')
plt.ylabel('Transaction Count')
plt.title('Transaction Counts by Hour of the Day')
plt.xticks(hourly_transaction_counts.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.annotate(f'Max: {most_transactions_count}', xy=(most_transactions_hour, most_transactions_count), xytext=(most_transactions_hour, most_transactions_count + 1),
             arrowprops=dict(facecolor='red', arrowstyle='->', linewidth=1),
             fontsize=12, color='red', ha='center')

plt.show()

print(f"Hour with the most transactions: {most_transactions_hour} (Total transactions: {most_transactions_count})")
```

```python
transactions_df['TransactionStartDateTime'] = pd.to_datetime(transactions_df['TransactionStartDateTime'])

transactions_df['TransactionHour'] = transactions_df['TransactionStartDateTime'].dt.hour

# Group transactions by hour and transaction type and count the number of transactions per hour
hourly_transaction_counts = transactions_df.groupby(['TransactionHour', 'TransactionTypeID']).size().unstack().fillna(0)


plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_transaction_counts, dashes=False)
plt.xlabel('Hour of the Day')
plt.ylabel('Transaction Count')
plt.title('Transaction Counts by Hour of the Day for Different Transaction Types')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.legend(title='Transaction Type', loc='upper right', labels=[f'Type {col}' for col in hourly_transaction_counts.columns])
plt.show()
```

```python
transactions_df['TransactionStartDateTime'] = pd.to_datetime(transactions_df['TransactionStartDateTime'])

transactions_df['TransactionHour'] = transactions_df['TransactionStartDateTime'].dt.hour

unique_location_bins = transactions_df['LocationBin'].unique()

# Create a line graph with time on the x-axis and separate lines for each location bin
plt.figure(figsize=(12, 6))
for location_bin in unique_location_bins:
    location_data = transactions_df[transactions_df['LocationBin'] == location_bin]
    hourly_location_counts = location_data.groupby('TransactionHour').size()
    sns.lineplot(x=hourly_location_counts.index, y=hourly_location_counts.values, label=location_bin)

plt.xlabel('Hour of the Day')
plt.ylabel('Transaction Count')
plt.title('Transaction Counts by Location Bin and Hour of the Day')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Location Bin')
plt.tight_layout()
plt.show()
```

```python
merged_data = pd.merge(transactions_df, customers_df, on='CardholderID')

# Create a heatmap for the relationship between customer occupations and transaction types
plt.figure(figsize=(12, 6))
sns.heatmap(data=merged_data.pivot_table(index='Occupation', columns='TransactionTypeID', values='TransactionAmount', aggfunc='size', fill_value=0), cmap='YlOrRd', annot=True, fmt='g')
plt.title('Occupation vs. Transaction Type')
plt.xlabel('Transaction Type')
plt.ylabel('Occupation')
plt.show()
```

```python
customers_df['Birth Date'] = pd.to_datetime(customers_df['Birth Date'])

customers_df['Age'] = (pd.Timestamp.now() - customers_df['Birth Date']).astype('<m8[Y]')

bins = [0, 18, 30, 45, 60, float('inf')]
labels = ['0-18', '19-30', '31-45', '46-60', '61+']


customers_df['AgeGroup'] = pd.cut(customers_df['Age'], bins=bins, labels=labels)

merged_data = pd.merge(transactions_df, customers_df, on='CardholderID')

# Group transactions by occupation and age group and sum the transaction amounts
occupation_age_transactions = merged_data.groupby(['Occupation', 'AgeGroup'])['TransactionAmount'].sum().unstack(fill_value=0)

plt.figure(figsize=(12, 6))
sns.heatmap(occupation_age_transactions, cmap='Oranges', annot=True, fmt='.2f', cbar=True)
plt.title('Occupation vs. Age Group')
plt.xlabel('Age Group')
plt.ylabel('Occupation')
plt.show()
```

```python
transactions_df['TransactionStartDateTime'] = pd.to_datetime(transactions_df['TransactionStartDateTime'])

transactions_df['Month'] = transactions_df['TransactionStartDateTime'].dt.month

monthly_trend = transactions_df.groupby('Month')['TransactionAmount'].sum() / 1000  # Scale to thousands

plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_trend.index, y=monthly_trend.values)
plt.title('Monthly Trend of Transaction Amount (in Thousands)')
plt.xlabel('Month')
plt.ylabel('Transaction Amount (in Thousands)')
plt.show()
```
