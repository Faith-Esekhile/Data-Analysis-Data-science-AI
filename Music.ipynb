```python
dummies=["Platform","Year","Genre","Publisher"]
```

```python
pd.get_dummies(df["Publisher"])
```

```python
def mapping (data,colunms):
    for x in colunms:
        dummies=pd.get_dummies(data[x])
        data=pd.concat([data,dummies],axis=1)
        data.drop(x,axis=1,inplace=True)
    return data
```

```python
df=mapping(df,dummies)
```
