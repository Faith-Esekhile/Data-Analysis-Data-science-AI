```python
# Gender & age
#calculated from client table decoding the 4th and 5th number, where if it's above 50 female otherwise male

client['gender'] = client.birth_number.astype(str).str[3:5].astype(int)
client['gender'] = np.where(client.gender > 50, 'female', 'male')

# Age
#Calculated from client table decoding, where age is derived from subtracting the first 2 numbers in datetime format by 2023
current_year = datetime.now().year
client['age'] = '19' + client.birth_number.astype(str).str[:2]
client.age = current_year - client.age.astype(int)
```

```python
# account_year
#To calculate the year the account was opened
account['account_year'] = account.date.astype(str).str[:4]
account.account_year = pd.to_datetime(account.account_year, format='%y%m').dt.strftime('%Y-%m')
account.head()
```

```python
#credit card years
# Merging account, disp and creditcard tables 
merge_df = pd.merge(account, disp, on='account_id')
merged = pd.merge(merge_df, card , on='disp_id')
merged
```

```python
# creditcard issue
merged['creditcard_issued'] = merged.issued.astype(str).str[:4]
merged.creditcard_issued = pd.to_datetime(merged.creditcard_issued, format='%y%m').dt.strftime('%Y-%m')
merged.head()
```

```python
#years to get credit card
merged['creditcard_issued'] = pd.to_datetime(merged['creditcard_issued'])
merged['account_year'] = pd.to_datetime(merged['account_year'])

merged['years_difference'] = (merged['creditcard_issued'] - merged['account_year']).dt.days // 365
merged.head()
```

```python
trans['date'] = pd.to_datetime(trans['date'])


trans['day_of_week'] = trans['date'].dt.dayofweek


def categorize_day(day):
    if day < 5:  
        return 'Weekday'
    else:        
        return 'Weekend'


trans['day_category'] = trans['day_of_week'].apply(categorize_day)

trans.head()
```

```python
trans['date'] = pd.to_datetime(trans['date'])

trans['year_month'] = trans['date'].dt.to_period('M')

average_monthly_transactions = trans.groupby('year_month').size().mean()

print("Average monthly transactions:", average_monthly_transactions)
```

```python
trans['date'] = pd.to_datetime(trans['date'])

max_date_rows = trans.loc[trans.groupby('account_id')['date'].idxmax()]

client_balances = max_date_rows[['account_id', 'balance_after_trans']]

print("Client Balances based on Maximum Transaction Date:")
print(client_balances)
```

```python
district = district.rename(columns={
    'A1': 'district_id',
    'A3': 'region'
})

district.head()
```

```python
average_salary = district.groupby('region')['A11'].mean().reset_index()

salary = average_salary.set_index('region')['A11']
district['average_salary'] = district.region.map(salary)
```

```python
average_unemployment = district.groupby('region')['A13'].mean().reset_index()

unemployment = average_unemployment.set_index('region')['A13']
district['average_unemployment'] = district.region.map(unemployment)
```

```python
average_crime = district.groupby('region')['A16'].mean().reset_index()

crime = average_crime.set_index('region')['A16']
district['average_crime'] = district.region.map(crime)
```

```python
district_column = ['A2', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A11', 'A12', 'A13', 'A14', 'A15', 'A16']

district = district.drop(district_column, axis=1)

district.head()
```

```python
merge1 = pd.merge(client, merged, on=['client_id', 'district_id'])

merge2 = pd.merge(merge1, loan, on='account_id')

merge3 = pd.merge(order, merge2, on='account_id')

merge4 = pd.merge(merge3, district, on='district_id')

dataset = pd.merge(merge4, trans, on='account_id')
dataset.head()
```

```python
dataset.date = pd.to_datetime(dataset.date).dt.strftime('%Y-%m')

avg_trans_month = dataset.groupby(['account_id', 'date'])['trans_amount'].sum().reset_index()

avg_trans_client = avg_trans_month.groupby('account_id')['trans_amount'].mean().reset_index()

client_b = avg_trans_client.set_index('account_id')['trans_amount']
dataset['avg_trans_permonth'] = round(dataset.account_id.map(client_b), 2)
```
